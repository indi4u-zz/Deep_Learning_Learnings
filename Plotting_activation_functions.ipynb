```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
i = np.arange(-5,15,0.5)
```

```python
## not used much due to vanishing/exploding gradient problems 

def sigmoid(x):
    return(1/ (1+ np.exp(-x)))
```

```python
sig_values = sigmoid(i)
```

```python
plt.plot(i,sig_values)
```

```python
## used for hidden layers 

def relu(x):
    return(np.maximum(x,0))
```

```python
relu_values = relu(i)
plt.plot(i,relu_values)
```

```python
## used in the output layer for classification problems
def softmax(x):
    return(np.exp(x) / np.sum(x,axis = 0))

softmax_values =softmax(i)
plt.plot(i,softmax_values)
```

```python
## not used much due to vanishing/exploding gradient problems 

def tanh(x):
    return(( 1 -  np.exp(-2*x) )/ (1+ np.exp(-2*x) ))
```

```python
#i = np.arange(-15,15,0.5)
plt.plot(i,tanh(i))
```
